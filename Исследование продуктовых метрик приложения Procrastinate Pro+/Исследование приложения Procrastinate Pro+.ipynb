## Функции для расчета и анализа LTV, ROI, удержания и конверсии

В первую очередь необходимо создать профили клиентов, то есть:

- Выделить дату, регион, канал привлечения, тип устройства первой сессии для каждого `user_id`.

- Обозначить, является ли пользователь платящим.

- Обозначить дату (без времени) первого посещения. Это необходимо для выделения когорт пользователей и вычисления лайтайма.

- Найти стоимость привлечения каждого клиента.

```python
def get_profiles(sessions, orders, ad_cost):
    
    # Выделение даты, канала привлечения, устройства и региона первого посещения
        profiles = (
            
            sessions
            .sort_values(by = ['user_id', 'session_start'])
            .groupby('user_id')
            .agg(
                {'session_start':'first',
                 'channel':'first',
                 'device':'first',
                 'region':'first'}
            )
            .rename(columns={'session_start':'first_ts'})
            .reset_index()
        )
        
        # Создание столбцов даты и месяца первой сессии пользователя
        profiles['dt'] = profiles['first_ts'].dt.date
        profiles['month'] = profiles['first_ts'].astype('datetime64[M]')
        
        # Обозначение платящего или не платящего пользователя
        profiles['payer'] = profiles['user_id'].isin(orders['user_id'].unique())
        
        # Вычисление стоимости привлечения каждого клиента
        # Подсчет общего количества пользователей в разрезе даты и канала привлечения
        new_users = (
            
            profiles.groupby(['dt', 'channel'])
            .agg({'user_id':'nunique'})
            .rename(columns = {'user_id':'unique_users'})
            .reset_index()
        
        )
        
        # Присоединение общего количества пользователей в разрезе даты и канала привлечения к 
        # таблице затрат на рекламу
        ad_cost = ad_cost.merge(new_users, on = ['dt', 'channel'], how = 'left')
        
        # Создание столбца стоимости каждого пользователя
        ad_cost['acquisition_cost'] = ad_cost['costs'] / ad_cost['unique_users']
        
        # Сопоставление профиля пользователя и его стоимости
        profiles = profiles.merge(ad_cost[['acquisition_cost', 'dt', 'channel']],
                                  on = ['dt', 'channel'],
                                  how = 'left'
                                 )
        # Заполнение null-значений, если пользователь является органическим
        profiles['acquisition_cost'] = profiles['acquisition_cost'].fillna(0)
       
        return profiles
```

Далее создадим функцию удержания клиентов. Для этого выполним следующие действия:

- Обозначим признак, по которому будем вычислять удержание клиентов (dimensions). Это может быть тип устройства, регион и т.д.

- Убираем из выборки пользователей, которые не успели "дожить" до момента анализа, если параметр `ignore_horizon = False`.

- Вычислим лайфтайм повторных посещений, присоединив к профилям пользователей таблицу их посещений.

- Создадим сводную таблицу удержания, в которой в качестве индекса будет выступать признак удержания клиентов (dimensions), в качестве столбцов лайфтайм, а в качестве значений количество уникальных пользователей за каждый лайфтайм с разбивкой по признаку (dimensions).

- Разделим количество уникальных пользователей на количество пользователей в когорте, тем самым вычислим Retention.

```python
# Функция создания таблицы удержания клиентов
def get_retention(profiles,
                  sessions,
                  observation_date,
                  horizon_days,
                  dimensions = [],
                  ignore_horizon = False
                 ):
    
    # Обознаение признаков группировки данных (признак платящего/не платящего клиента проставляется по умолчанию)
    dimensions = ['payer'] + dimensions
    
    # Вычисление горизонта анализа исходя из момента анализа
    last_suitable_acquisition_date = observation_date
    
    # Если ignore_horizon = False, исключаем пользователей, "не доживших" до горизонта анализа
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days = horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')
    
    # Присоединение таблицы посещений к сырым данным для подсчета удержания пользователей
    result_raw = result_raw.merge(
        sessions[['user_id', 'session_start']],
        on = 'user_id',
        how = 'left'
    )
    
    # Подсчет лайфтайма
    result_raw['lifetime'] = (
        result_raw['session_start'] - result_raw['first_ts']
    ).dt.days
    
    # Реализация таблицы удержания, исходя из желаемого признака dims
    def group_by_dimensions(df, dims, horizon_days):
        result = df.pivot_table(
            index = dims,
            columns = 'lifetime',
            values = 'user_id',
            aggfunc = 'nunique'
        )
        
        cohort_size = result[0]
        # Деление каждой ячейки на размер когорт
        result = result.div(cohort_size, axis=0)
        result['cohort_size'] = cohort_size
        # Оставляем в итоговой выборке только те лайфтаймы, которые меньше горизонта анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        
        return result
    
    # Создание таблицы удержания и таблицы удержания в диманике
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)
    result_in_time = group_by_dimensions(result_raw, dimensions + ['dt'], horizon_days)
    
    return result_raw, result_grouped, result_in_time
```

Теперь создадим функцию вычисления конверсии первой покупки. Для этого:

- Обозначим признак, по которому будем вычислять конверсию первой покупки (dimensions).

- Убираем из выборки пользователей, которые не успели "дожить" до момента анализа, если `ignore_horizon = False`.

- Из таблицы покупок обозначим дату первой покупки.

- Вычислим лайфтайм, с помощью присоединения к профилям пользователей даты их первой покупки.

- Создадим сводную таблицу конверсии, в которой в качестве индекса будет выступать признак конверсии клиентов (dimensions), в качестве столбцов лайфтайм, а в качестве значений количество уникальных пользователей за каждый лайфтайм с разбивкой по признаку (dimensions).

- Разделим количество уникальных пользователей, совершивших первую покупку, на количество пользователей в когорте, тем самым вычислим конверсию первой покупки.

```python
# Функция создания таблицы конверсии в первую покупку
def get_conversion(profiles,
                   purchases,
                   observation_date,
                   horizon_days,
                   dimensions=[],
                   ignore_horizon=False):
    
    # Вычисление горизонта анализа, исходя из момента анализа
    last_suitable_acquisition_date = observation_date
    
    # Исключаем пользователей, "не доживших" до горизонта анализа
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days = horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')
    
    # Создание вспомогательной таблицы даты первой покупки каждого клиента
    first_purchases = (purchases.sort_values(by = ['user_id', 'event_dt'])
                       .groupby('user_id')
                       .agg({'event_dt':'first'})
                       .reset_index()
                      )
    
    # Присоединение таблицы первого посещения и первой покупки для расчета конверсии
    result_raw = result_raw.merge(first_purchases[['user_id', 'event_dt']],
                                  on = 'user_id',
                                  how = 'left'
                                 )
    # Вычисление лайфтайма
    result_raw['lifetime'] = (result_raw['event_dt'] - result_raw['first_ts']).dt.days
    
    # Если в условиях группировки (dimensions) ничего нет,
    # тогда создаем костыльное поле (cohort) для группировки
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users' 
        dimensions = dimensions + ['cohort']
    
    # Реализация таблицы конверсии в первую покупку
    def group_by_dimensions(df, dims, horizon_days):
        result = df.pivot_table(index = dims,
                                columns = 'lifetime',
                                values = 'user_id',
                                aggfunc = 'nunique')
        
        # Вычисление накопительной суммы уникальных пользователей, совершивших первую покупку
        result = result.fillna(0).cumsum(axis = 1)
        
        # Вычисление размеров когорт
        cohort_size = (
            df.groupby(dims)
            .agg({'user_id':'nunique'})
            .rename(columns={'user_id':'cohort_size'})
        )
        result = cohort_size.merge(result, on = dims, how = 'left').fillna(0)
        
        # Деление каждой ячейки на размер когорт
        result = result.div(result['cohort_size'], axis = 0)
        # Оставляем в итоговой выборке только те лайфтаймы, которые меньше горизонта анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        result['cohort_size'] = cohort_size
        return result
    
    # Создание таблицы конверсии в первую покупку
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)
    result_in_time = group_by_dimensions(result_raw, dimensions + ['dt'], horizon_days)
    
    return result_raw, result_grouped, result_in_time
```

Теперь создадим функцию вычисления LTV, LTV в динамике, CAC, ROI и ROI в динамике. LTV и ROI в динамике в дальнейшем будем вычислять за 14-ый лайфтайм каждой когорты, так как по условию пользователь должен окупаться не позднее, чем через 14 дней после привлечения.

Для создания таблицы LTV необходимо:

- К таблице профилей присоединить таблицу покупок по каждому клиенту.

- Рассчитать  лайфтам от первого посещения до покупки.

- Создать сводную таблицу "пожизненной" ценности клиента, в которой в качестве индекса будет выступать признак LTV (dimensions), в качестве столбцов лайфтайм, а в качестве значений сумма, потраченная пользователями за каждый лайфтайм с разбивкой по признаку (dimensions).

Для создания таблицы CAC необходимо вычислить средние затраты на рекламу в зависимости от признака dimensions.

Для создания таблицы ROI необходимо разделить все значения лайфтаймов таблицы LTV на соответствующий CAC.

Таблицы LTV и ROI в динамике  рассчитываются аналогично, только в dimensions добавляется поле даты первого посещения.

```python
# Создание функции расчета ltv, cac и roi
def get_ltv(
    profiles,
    purchases,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False
):
    
    # Вычисление горизонта анализа, исходя из момента анализа
    last_suitable_acquisition_date = observation_date
    
    # Исключаем пользователей, "не доживших" до горизонта анализа
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days = horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')
    
    # Присоединение таблицы покупок к сырым данным для расчета ltv
    result_raw = result_raw.merge(
        purchases[['user_id', 'event_dt', 'revenue']], on = 'user_id', how = 'left'
    )    
    # Расчет лайфтама
    result_raw['lifetime'] = (result_raw['event_dt'] - result_raw['first_ts']).dt.days
    
    # Если в условиях группировки (dimensions) ничего нет,
    # тогда создаем костыльное поле (cohort) для группировки
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users'
        dimensions = dimensions + ['cohort'] 
    
    
    # Реализация таблиц ltv, cac и roi
    def group_by_dimensions(df, dims, horizon_days):
        
        # Реализация таблицы ltv
        result = df.pivot_table(
            index = dims,
            columns = 'lifetime',
            values = 'revenue',
            aggfunc = 'sum'
        )
        
        # Вычисляем сумму с накоплением
        result = result.fillna(0).cumsum(axis=1)
        
        # Вычисляем размер когорт
        cohort_size = (df.groupby(dims)
                       .agg({'user_id':'nunique'})
                       .rename(columns = {'user_id': 'cohort_size'})
                      )
        result = cohort_size.merge(result, on = dims, how = 'left').fillna(0)
        
        # Делим каждую ячейку суммы с накоплением на размер соответствующей когорты
        result = result.div(result['cohort_size'], axis = 0)
        # Оставляем в итоговой выборке только те лайфтаймы, которые меньше горизонта анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        result['cohort_size'] = cohort_size
        
        # Реализация таблицы cac
        cac = df[['user_id', 'acquisition_cost'] + dims].drop_duplicates()
        cac = (
            cac.groupby(dims)
            .agg({'acquisition_cost':'mean'})
            .rename(columns = {'acquisition_cost':'cac'})
        )

        # Реализация таблицы roi
        
        # Делим каждую ячейку суммы с накоплением на CAC
        roi = result.div(cac['cac'], axis = 0)
        # Удаляем все значения бесконечности, которые образовались при делении на ноль
        roi = roi[~roi['cohort_size'].isin([np.inf])]
        roi['cohort_size'] = cohort_size
        roi['cac'] = cac['cac']
        # Оставляем в итоговой выборке только те лайфтаймы, которые меньше горизонта анализа
        roi = roi[['cohort_size', 'cac'] + list(range(horizon_days))]
        
        return result, roi
    
    # Создание таблиц ltv и roi
    result_grouped, roi_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)
    
     
    if 'cohort' in dimensions:
        dimensions = []
        
    # Создание таблиц ltv и roi в динамике
    result_in_time, roi_in_time = group_by_dimensions(result_raw, dimensions + ['dt'], horizon_days)
    
    return (
        result_raw,
        result_grouped,
        result_in_time,
        roi_grouped,
        roi_in_time
    )
```

Далее реализуем функции построения графиков retention, conversion, LTV и ROI.

Так как в изначальных данных представлен очень обширный срок наблюдения (более полугода), то необходимо сделать графики LTV и ROI в динамике менее "шумными". Для этого создадим функцию `filter_data`, которая будет реализовывать механизм скользящего среднего для каждого столбца.

```python
# Функция вычисления скользящего среднего для сглаживания графиков
def filter_data(df, window):
    
    # Для каждого столбца вычисляем скользящее среднее по указанному окну
    for column in df.columns.values:
        df[column] = df[column].rolling(window).mean()
        
    return df
```

В функции `plot_retention` будут реализовываться следующие графики:

- Удержание платящих клиентов.

- Удержание не платящих клиентов.

- Динамика удержания платящих пользователей на n-й день.

- Динамика удержания не платящих пользователей на n-й день.

```python
# Функция создния графиков удержания
def plot_retention(retention, retention_history, horizon, window=7):
    
    # Задаем размер сетки для графиков
    plt.figure(figsize = (15, 10))
    
    # Исключаем из таблиц вспомогательные поля, которые могу повлиять на график
    retention = retention.drop(columns=['cohort_size', 0])
    retention_history = retention_history.drop(columns=['cohort_size'])[
        [horizon - 1]]
    
    # Если в индексах таблицы удержания только payer,
    # добавляем второй признак — cohort
    if retention.index.nlevels == 1:
        retention['cohort'] = 'All users'
        retention = retention.reset_index().set_index(['cohort', 'payer'])
    
    # Обозначаем, что в графике будет 2 строчки и 2 столбца
    ax1 = plt.subplot(2,2,1)
    
    # В 1 ячейке строим график удержания платящих клиентов
    retention.query('payer == True').droplevel('payer').T.plot(grid=True, ax=ax1)   
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Удержание платящих пользователей')
    
    # В 2 ячейке строим график удержания неплатящих клиентов
    ax2 = plt.subplot(2,2,2,sharey=ax1)
    retention.query('payer == False').droplevel('payer').T.plot(grid = True, ax=ax2)
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Удержание не платящих пользователей')
    
    # В 3 ячейке динамика удержания платящих клиентов
    ax3 = plt.subplot(2,2,3)
    columns = [
        
        name for name in retention_history.index.names if name not in ['dt', 'payer']
    ]
    filtered_data = (
        retention_history.query('payer == True')
        .pivot_table(index = 'dt',
                     columns=columns,
                     values=horizon - 1,
                     aggfunc='mean')
    )
    
    # Сглаживаем значения с помощью функции filter_data и строим график
    filter_data(filtered_data, window).plot(grid=True, ax=ax3)
    plt.xlabel('Дата привлечения')
    plt.title(
        'Динамика удержания платящих пользователей на {}-й день'.format(
            horizon
        )
    )
    
    # В 4 ячейке динамика удержания неплатящих клиентов
    ax4 = plt.subplot(2,2,4,sharey=ax3)
    filtered_data = (
        retention_history.query('payer == False')
        .pivot_table(index = 'dt',
                     columns = columns,
                     values=horizon - 1,
                     aggfunc='mean')
    )
        
    filter_data(filtered_data, window).plot(grid=True, ax=ax4)
    plt.xlabel('Дата привлечения')
    plt.title(
        'Динамика удержания не платящих пользователей на {}-й день'.format(
            horizon
        )
    )
    
    plt.tight_layout()
    plt.show()
```

В функции `plot_conversion` будут реализовываться следующие графики:

- Конверсия в первую покупку.

- Динамика конверсии на n-ый день.

```python
# Функция создния графиков конверсии
def plot_conversion(conversion,
                    conversion_history,
                    horizon,
                    window=7):
    
    # Задаем размер сетки для графиков
    plt.figure(figsize = (15, 10))
    
    # Исключаем из таблиц вспомогательные поля, которые могу повлиять на график
    conversion = conversion.drop(columns=['cohort_size'])
    conversion_history = conversion_history.drop(columns=['cohort_size'])[[horizon - 1]]

    # В 1 ячейке конверсия в первую покупку
    ax1 = plt.subplot(1,2,1)
    conversion.T.plot(grid=True, ax=ax1)
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Конверсия пользователей')
    
    # В 2 ячейке конверсия в первую покупку в динамике
    ax2 = plt.subplot(1,2,2, sharey=ax1)
    columns = [
        # Столбцы сводной таблицы - все столбцы индекса, кроме даты
        # Значения сводной таблицы - среднее значение конверсии за определенный день (horizon)
        name for name in conversion_history.index.names if name not in ['dt']
    ]
    
    filtered_data = conversion_history.pivot_table(index = 'dt',
                                                   columns = columns,
                                                   values = horizon - 1,
                                                   aggfunc='mean')
    
    # Сглаживаем значения с помощью функции filter_data и строим график
    filter_data(filtered_data, window).plot(grid=True, ax = ax2)
    
    plt.xlabel('Дата привлечения')
    plt.title('Динамика конверсии пользователей на {}-й день'.format(horizon))
    
    plt.tight_layout()
    plt.show()
```

В функции `plot_ltv_roi` будут реализовываться следующие графики:

- График LTV.

- Динамика LTV пользователей на n-й день.

- Динамика стоимости привлечения пользователей (CAC).

- График ROI.

- Динамика ROI пользователей на n-й день. 

```python
def plot_ltv_roi(ltv,
                 ltv_history,
                 roi,
                 roi_history,
                 horizon,
                 window=7):
    
    # Задаем размер сетки для графиков
    plt.figure(figsize=(20,10))
    
    # Исключаем из таблиц вспомогательные поля, которые могу повлиять на график
    ltv = ltv.drop(columns=['cohort_size'])
    ltv_history = ltv_history.drop(columns=['cohort_size'])[[horizon - 1]]
    
    # Создаем фрейм затрат на маркетинг
    cac_history = roi_history[['cac']]
    
    # Исключаем из таблиц вспомогательные поля, которые могу повлиять на график
    roi = roi.drop(columns=['cohort_size', 'cac'])
    roi_history = roi_history.drop(columns=['cohort_size', 'cac'])[[horizon - 1]]
    
    # В 1 ячейке график LTV
    ax1 = plt.subplot(2,3,1)
    ltv.T.plot(grid = True, ax = ax1)
    plt.legend()
    plt.title('LTV')
    plt.xlabel('Лайфтайм')
    
    # В 2 ячейке график LTV в динамике
    ax2 = plt.subplot(2,3,2,sharey=ax1)
    
    columns = [
        # столбцы сводной таблицы - все столбцы индекса, кроме даты
        # значения сводной таблицы - средние значения LTV за определенный день (horizon)
        name for name in ltv_history.index.names if name not in ['dt']
    ]
    filtered_data = ltv_history.pivot_table(index = 'dt',
                                            columns = columns,
                                            values = horizon - 1,
                                            aggfunc = 'mean')
    
    # Сглаживаем значения с помощью функции filter_data и строим график
    filter_data(filtered_data, window).plot(grid=True, ax=ax2)
    plt.xlabel('Дата привлечения')
    plt.title('Динамика LTV пользователей на {}-й день'.format(horizon))
    
    # В 3 ячейке график затрат на маркетинг в динамике
    ax3 = plt.subplot(2, 3, 3, sharey=ax1)
    columns = [
        # столбцы сводной таблицы - все столбцы индекса, кроме даты
        # значения сводной таблицы - средние значения cac за определенный день (horizon)
        name for name in cac_history.index.names if name not in ['dt']
    ]
    filtered_data = cac_history.pivot_table(
        index='dt', columns=columns, values='cac', aggfunc='mean'
    )
    # Сглаживаем значения с помощью функции filter_data и строим график
    filter_data(filtered_data, window).plot(grid = True, ax=ax3)
    plt.xlabel('Дата привлечения')
    plt.title('Динамика стоимости привлечения пользователей')
    
    # В 4 ячейке график ROI
    ax4 = plt.subplot(2,3,4)
    roi.T.plot(grid=True, ax=ax4)
    # Вспомогательная линия для обозначения границы окупаемости затрат на маркетинг
    plt.axhline(y=1,color='red', linestyle = '--', label = 'Уровень окупаемости')
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('ROI')
    
    # В 5 ячейке график ROI в динамике
    ax5 = plt.subplot(2,3,5,sharey=ax4)
    columns = [
        # столбцы сводной таблицы - все столбцы индекса, кроме даты
        # значения сводной таблицы - средние значения roi за определенный день (horizon)
        name for name in roi_history.index.names if name not in ['dt']
    ]
    filtered_data = roi_history.pivot_table(index='dt', columns=columns, values=horizon - 1, aggfunc='mean')
    
    # Сглаживаем значения с помощью функции filter_data и строим график
    filter_data(filtered_data, window).plot(grid=True, ax=ax5)
    plt.axhline(y=1, color='red', linestyle='--', label='Уровень окупаемости')
    plt.xlabel('Дата привлечения')
    plt.title('Динамика ROI пользователей на {}-й день'.format(horizon))
    
    plt.tight_layout()
    plt.show()
```

В начале мая на США и так выделялось больше рекламных средств, чем на другие страны. Затраты на рекламу в США в мае окупались. Поэтому действия маркетологов были весьма логичны.

**Что же было в США после первоначального вложения средств:**

Прибыль в США действительно начала расти, но к середине июня она достигла своего пика и начала снижаться. На пике прибыли удалось добиться небольшой окупаемости. Однако в других странах окупаемость была намного выше.

В конце июня маркетологи решают вложить еще больше средств в TipTop, а на FaceBoom выделить ту же сумму, что и в конце мая. Это снова дает свои плоды и прибыль снова увеличивается, достигая пика в середине июля. Пик в середине июля и в середине июня равны, при больших затратах на рекламу в июле. Следовательно, с конца июня рекламная кампания в США перестает окупаться.

В конце июля прибыль снова падает, и маркетологи повторяют уже знакомые действия: вкладывают еще больше средств в TipTop, рассчитывая выйти на новый уровень окупаемости. В начале августа прибыль начинает расти, но она все равно не окупает затрат на привлечение пользователей.

Такая же ситуация происходит в сентябре и октябре: всё бОльшие средства вкладываются в TipTop и примерно постоянная сумма идет на FaceBoom, но это не приносит никакой окупаемости.

**Какие ошибки были допущены:**

Изначально маркетологи сделали ставку на рынок США. Стратегия состояла в том, чтобы привлечь пользователей, которые "не отваливаются" после первых покупок. На мой взгляд, в этом и была главная ошибка. Стоило распределить рекламные деньги на США, а также на страны Европы, в которых при меньшей конверсии наблюдается большее удержание. При чем, на Европу, в особенности на Германию и Великобританию, выделить больше средств, чем на США. При прочих равных условиях, мы привлекли бы больше новых пользователей из Европы и повысили бы конверсию в первую покупку, а значит был бы хороший шанс, что новые пользователи остались бы с нами надолго и приносили бы прибыль в дальнейшем.

Среди европейских источников привлечения можно было бы вложиться в LeapBob и lambdaMediaAds, так как у них была наибольшая окупаемость в мае.

Постфактум можно сказать, что необходимо выделить ресурсы на дополнительное исследование логов действий пользователей в приложении, чтобы выяснить, по какой причине пользователи в США, да и в Европе, быстро "отваливаются" после совершения первой покупки.
