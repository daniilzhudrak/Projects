## Предобработка данных

### Обработка пропусков

#### Устранение пропусков в столбце balcony

При первичном анализе было выявлено, что в столбце `balcony` содержится большое количество пропусков (11519). Предположим, что они связаны с тем, что у людей в принципе нет балкона и они не стали указывать эту информацию в соответствующей графе. Исходя из этого утверждения, изменим все пропущенные значения в столбце balcony на значение 0.

```python
df.loc[df['balcony'].isna(), 'balcony'] = 0 #замена пропусков на значения 0 
```

```python
df['balcony'].isna().sum() #проверка устранения пропусков
```

#### Устранение пропусков в столбце is_apartment

Аналогичный вывод можно сделать и о столбце `is_apartment`. Этот столбец содержит огромное количество пропусков и скорее всего они вызваны тем, что недвижимость большинства клиентов не является апартаментами. Из-за этого люди предпочитают не заполнять эту информацию. Поэтому заменим пропуски в данном столбце значением False.

```python
df.loc[df['is_apartment'].isna(), 'is_apartment'] = False #замена пропусков на значние False
```

```python
df['is_apartment'].isna().sum() #проверка замены пропусков
```

Данные столбцы можно и не изменять. Тип `float` не делает какие-либо арифметические операции с этими столбцами невозможными, но визуально тип int воспринимается легче.

Для выявления дня недели из даты публикации объявления о продаже недвижимости применим атрибут `dt` и к нему метод `weekday`. Данный метод поможет выявить только номер дня недели (0 - понедельник, 6 - воскресенье). Безусловно, это неудобно для дальнейшего анализа, поэтому создадим функцию, которая будет преобразовывать номер дня недели в словесный вариант.

```python
#функция преобразования номера дня недели
def day_of_the_week(day):
    if day == 0:
        return 'Понедельник'
    if day == 1:
        return 'Вторник'
    if day == 2:
        return 'Среда'
    if day == 3:
        return 'Четверг'
    if day == 4:
        return 'Пятница'
    if day == 5:
        return 'Суббота'
    if day == 6:
        return 'Воскресенье'
```

```python
#создание нового столбца weekday_number
df['weekday_number'] = df['first_day_exposition'].dt.weekday
```

Далее с помощью метода apply применим функцию к созданному столбцу `weekday_number`.

```python
#примененение функции day_of_the_week и создание нового столбца weekday
df['weekday'] = df['weekday_number'].apply(day_of_the_week)
```

На данном этапе можно сразу произвести анализ поля `weekday` и выяснить в какие дни люди чаще всего выкладывают объявления о продаже недвижимости.

Это объяснимо, так как в большинстве случаев люди рассчитывают на быструю продажу и, выкладывая объявления среди недели, они планируют показ квартиры для потенциальных покупателей на выходные. ### Категорирование данных по этажу

Произведем категорирование данных по этажу, создав следующие группы:

- 1 (если этаж первый);

- 2 (если этаж последний);

- 3 этаж в доме (если этаж первый и при этом последний);

- 4 (если этаж не первый и не последний).

Для этого создадим соответствующую функцию floor_definition.

```python
def floor_definition(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1 and floors_total != floor:
        return 1
    if floor == floors_total and floor != 1:
        return 2
    if floor == 1 and floors_total == floor:
        return 3
    return 4
```

Для того чтобы функция работала корректно, ей необходимо передать несколько столбцов каждой строки (floor и floors_total). Для этого в методе `apply` применим параметр `axis = 1`.

```python
#содание нового столбца floor_group
df['floor_group'] = df.apply(floor_definition, axis = 1)
```

```python
df['floor_group'].value_counts()
```

Выполним анализ созданного столбца `floor_group`.

### Построение гистограммы для каждого параметра

Выведем гистограммы и произведем их анализ для следующих параметров:

- площадь.
- цена.
- число комнат.
- высота потолков.

Из пункта 5.1 можно сделать вывод, что в каждом перечисленном столбце присутствуют выбивающиеся значения и для корректного анализа будет лучше оставить их вне рассматриваемого диапазона с помощью параметра `range`. Определять выбивающиеся значения будем с помощью диаграммы размаха, то есть будем считать выбросом все значения, которые меньше `Q1 - 1,5*IQR` и больше `Q3 + 1,5*IQR`.

Для оптимизации процесса построения графиков создадим функцию `plotting`, которая примет на вход 2 параметра. Первый параметр - список столбцов, по которым необходимо построить гистограммы. Второй параметр - название датафрейма, в котором содержатся эти столбцы (по умолчанию этот параметр равен `df`, так как он используется чаще всего, но в дальнейшем будет необходимо построить гистограммы и по другим датафреймам).

```python
#импорт модуля pyplot из библиотеки matplotlib для вывода гистограмм и обозначения названия графика
import matplotlib.pyplot as plt
```

```python
#Функция для построения графика
def plotting(coloumn_list, dataframe = df):
    for i in coloumn_list:
        Q1 = dataframe[i].quantile(0.25) #Вычисление 1-го квартиля 
        Q3 = dataframe[i].quantile(0.75) #Вычисление 3-го квартиля
        left =  Q1 - 1.5 * (Q3 - Q1) #Определение левой границы для параметра range
        right = Q3 + 1.5 * (Q3 - Q1) #Определение правой границы для параметра range
        if left < 0:
            left = 0 #В случае если левая граница меньше 0, то присвоим ей значение 0, так будет легче чиатать график
        dataframe[i].hist(bins = 50, range = (left, right), figsize=(10,7))
        plt.title(f'Гистограмма столбца {i}')
        plt.show()
```

```python
coloumn_list = ['total_area', 'last_price', 'rooms', 'ceiling_height']
```

```python
plotting(coloumn_list)
```

Далее с помощью метода `query` реализуем срез данных только по Санкт-Петербургу и сгруппируем созданный срез по столбцу `city_center_nearest_in_km`, записав данные в отдельный датафрейм `price_per_km_in_spb`. На основе этого датафрейма построим график.
